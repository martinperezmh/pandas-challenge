```python
import pandas as pd
from pathlib import Path

file=Path("Resources/schools_complete.csv")
file1=Path("Resources/students_complete.csv")

schools_df=pd.read_csv(file)

students_df=pd.read_csv(file1)
```

```python
list_schools_district=schools_df["school_name"].unique()


total_schools_district=len(list_schools_district)

total_students_district=schools_df["size"].sum()

total_budget_district=schools_df["budget"].sum()

district_math_avg=students_df["math_score"].mean()

district_reading_avg=students_df["reading_score"].mean()

num_pass_math_district=students_df[(students_df["math_score"]>=70)].count()["student_name"]
percent_pass_math_district=((num_pass_math_district)/(total_students_district))*100

num_pass_reading_district=students_df[(students_df["reading_score"]>=70)].count()["student_name"]
percent_pass_reading_district=((num_pass_reading_district)/(total_students_district))*100

overall_num_pass_math_reading=students_df[(students_df["math_score"]>=70)&(students_df["reading_score"]>=70)].count()["student_name"]
overall_percent_pass_math_reading=(overall_num_pass_math_reading/total_students_district)*100


district_summary = pd.DataFrame({
    "Total Schools":[total_schools_district],"Total Students":[total_students_district],"Total Budget": [total_budget_district],"Average Math Score":[round(district_math_avg,2)],"Average Reading Score": [round(district_reading_avg,2)],"% Passing Math": [round(percent_pass_math_district,2)],"% Passing Reading": [round(percent_pass_reading_district,2)], "% Overall Passing":[round(overall_percent_pass_math_reading,2)]})

district_summary.head()
```

```python
school_district_database=pd.merge(students_df,schools_df,on="school_name",how="right")
students_per_school = school_district_database["school_name"].value_counts()

per_school_grouping=school_district_database.groupby(["school_name"])

school_names=per_school_grouping["school_name"].unique()

school_types=schools_df.set_index('school_name')['type']


per_school_budget=schools_df.set_index('school_name')["budget"]
per_school_capita=(per_school_budget/students_per_school)

school_avg_math=per_school_grouping["math_score"].mean()
school_avg_reading=per_school_grouping["reading_score"].mean()

students_passing_math=school_district_database[(school_district_database["math_score"]>=70)].groupby("school_name")['Student ID'].count()
students_passing_reading=school_district_database[(school_district_database["reading_score"]>=70)].groupby("school_name")['Student ID'].count()
students_passing_math_reading=school_district_database[(school_district_database["math_score"]>=70)&(school_district_database["reading_score"]>=70)].groupby("school_name")['Student ID'].count()

per_school_passing_math= (students_passing_math/(students_per_school)*100)
per_school_passing_reading= (students_passing_reading/(students_per_school)*100)
per_school_overall_passing= (students_passing_math_reading/(students_per_school)*100)

per_school_summary=pd.DataFrame({
    "type":school_types,
    "size":students_per_school,
    "budget":per_school_budget,
    "Per Student Budget":per_school_capita,
    "Average Math Score":round(school_avg_math,2),
    "Average Reading Score":round(school_avg_reading,2),
    "% Passing Math":round(per_school_passing_math,2),
    "% Passing Reading":round(per_school_passing_reading,2),
    "% Overall Passing":round(per_school_overall_passing,2)
                                })


per_school_summary=per_school_summary.rename(columns={"type":"School Type","size":"Total Students","budget":"Total School Budget"})

per_school_summary['Total School Budget'] = per_school_summary['Total School Budget'].map('${:,.2f}'.format)
per_school_summary['Per Student Budget'] = per_school_summary['Per Student Budget'].map('${:,.2f}'.format)

per_school_summary
```

```python
ninth_grade_reading_scores=school_district_database.loc[school_district_database['grade']=='9th'].groupby('school_name')["reading_score"].mean()

tenth_grade_reading_scores=school_district_database.loc[school_district_database['grade']=='10th'].groupby('school_name')["reading_score"].mean()
eleventh_grade_reading_scores=school_district_database.loc[school_district_database['grade']=='11th'].groupby('school_name')["reading_score"].mean()
twelfth_grade_reading_scores=school_district_database.loc[school_district_database['grade']=='12th'].groupby('school_name')["reading_score"].mean()

reading_summary=pd.DataFrame({
    "9th":round(ninth_grade_reading_scores,2),
    "10th":round(tenth_grade_reading_scores,2),
    "11th":round(eleventh_grade_reading_scores,2),
    "12th":round(twelfth_grade_reading_scores,2)
})

reading_summary=reading_summary.rename(columns={"school_name":""})
reading_summary.head()
```

```python
spending_bins = [0, 585, 630, 645, 680]
spending_labels = ["<$585", "$585-630", "$630-645", "$645-680"]

school_spend_summary=per_school_summary.copy()

school_spend_summary['Per Student Budget']=school_spend_summary['Per Student Budget'].str.replace('$','',regex=False)
school_spend_summary['Per Student Budget']=school_spend_summary['Per Student Budget'].astype(float)

school_spend_summary["Spending Ranges (Per Student)"]=pd.cut(school_spend_summary['Per Student Budget'],spending_bins,labels=spending_labels)

spending_math_scores = school_spend_summary.groupby(["Spending Ranges (Per Student)"])["Average Math Score"].mean()
spending_reading_scores = school_spend_summary.groupby(["Spending Ranges (Per Student)"])["Average Reading Score"].mean()
spending_passing_math = school_spend_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Math"].mean()
spending_passing_reading = school_spend_summary.groupby(["Spending Ranges (Per Student)"])["% Passing Reading"].mean()
overall_passing_spending = school_spend_summary.groupby(["Spending Ranges (Per Student)"])["% Overall Passing"].mean()

spend_summary=pd.DataFrame({
    "Average Math Score":spending_math_scores,
    "Average Reading Score": spending_reading_scores,
    "% Passing Math": spending_passing_math,
    "% Passing Reading": spending_passing_reading,
    "% Overall Passing": overall_passing_spending,
})

spend_summary
```

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]

school_size_summary= per_school_summary.copy()
school_size_summary["School Size"]=pd.cut(school_size_summary['Total Students'],size_bins,labels=size_labels)

size_math_scores = school_size_summary.groupby(["School Size"])["Average Math Score"].mean()
size_reading_scores = school_size_summary.groupby(["School Size"])["Average Reading Score"].mean()
size_passing_math = school_size_summary.groupby(["School Size"])["% Passing Math"].mean()
size_passing_reading = school_size_summary.groupby(["School Size"])["% Passing Reading"].mean()
size_overall_passing = school_size_summary.groupby(["School Size"])["% Overall Passing"].mean()

size_summary=pd.DataFrame({
    "Average Math Score":size_math_scores,
    "Average Reading Score": size_reading_scores,
    "% Passing Math": size_passing_math,
    "% Passing Reading": size_passing_reading,
    "% Overall Passing": size_overall_passing,
})

size_summary
```
